Train and test

```python
from sklearn.model_selection import train_test_split

np.random.seed(0)
df_train, df_test = train_test_split(data_lr, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
num_vars = ['wheelbase', 'curbweight', 'enginesize', 'boreratio', 'horsepower','carlength','carwidth','price']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
plt.figure(figsize=(35,25))
sns.heatmap(df_train.corr(),annot=True)
plt.show()
```

```python
Y_train = df_train.pop('price')
X_train = df_train
```

Model Building

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
lm = LinearRegression()
lm.fit(X_train,Y_train)
rfe=RFE(lm)
rfe= rfe.fit(X_train, Y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
X_train.columns[rfe.support_]
```

```python
X_train_rfe = X_train[X_train.columns[rfe.support_]]
X_train_rfe.head()
```

```python
def build_model(X,Y):
    X = sm.add_constant(X)
    lm = sm.OLS(Y,X).fit()
    print(lm.summary())
    return X

def checkVIF(X):
    vif = pd.DataFrame()
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return(vif)
```

```python
X_train_new = build_model(X_train_rfe,Y_train)
```

```python
X_train_new = build_model(X_train_new,Y_train)
```

```python
checkVIF(X_train_new)
```

```python
lm = sm.OLS(Y_train,X_train_new).fit()
Y_train_price = lm.predict(X_train_new)
```

```python
fig = plt.figure()
sns.histplot((Y_train - Y_train_price), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)
plt.xlabel('Errors', fontsize = 18)
```

Prediction and Evaluation

```python
num_vars = ['wheelbase', 'curbweight', 'enginesize', 'boreratio', 'horsepower','carlength','carwidth','price']
df_test[num_vars] = scaler.fit_transform(df_test[num_vars])
```

```python
Y_test = df_test.pop('price')
X_test = df_test
```

```python
X_train_new = X_train_new.drop('const',axis=1)
X_test_new = X_test[X_train_new.columns]

X_test_new = sm.add_constant(X_test_new)
```

```python
Y_pred = lm.predict(X_test_new)
```

```python
from sklearn.metrics import r2_score
r2_score(Y_test, Y_pred)
```
